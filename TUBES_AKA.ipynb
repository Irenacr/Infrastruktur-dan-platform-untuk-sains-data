```python
# Install chardet (jika belum)
!pip install chardet

import pandas as pd
import chardet

# Membaca data dari file CSV (menangani error parsing)
try:
    netflix_data = pd.read_csv('/content/NetflixOriginals.csv', encoding='latin1', on_bad_lines='skip', quoting=1)  # quoting=1 untuk menangani masalah kutipan
    demographic_data = pd.read_csv('/content/demographic.csv', encoding='latin1', on_bad_lines='skip', quoting=1)
    filmtv_data = pd.read_csv('/content/filmtv_movies.csv', encoding='latin1', on_bad_lines='skip', quoting=1)
except pd.errors.ParserError as e:
    print("Error saat membaca file CSV:", e)

# Check if the 'genre' or 'genres' column exists in filmtv_data
if 'genre' not in filmtv_data.columns and 'genres' not in filmtv_data.columns:
    filmtv_data['genre'] = 'Unknown'  # or any suitable default value
elif 'genres' in filmtv_data.columns:
    # Rename it to 'genre' if 'genres' exist
    filmtv_data.rename(columns={'genres': 'genre'}, inplace=True)

# Rename the 'Genre' column in netflix_data to 'genre'
if 'Genre' in netflix_data.columns:
    netflix_data.rename(columns={'Genre': 'genre'}, inplace=True)

# Rename the 'Title' column in netflix_data to 'title'
if 'Title' in netflix_data.columns:
    netflix_data.rename(columns={'Title': 'title'}, inplace=True)

# Rename the 'genres' column in demographic_data to 'genre'
if 'genres' in demographic_data.columns:
    demographic_data.rename(columns={'genres': 'genre'}, inplace=True)

# Rename the 'Title' column in demographic_data to 'title'
if 'Title' in demographic_data.columns:
    demographic_data.rename(columns={'Title': 'title'}, inplace=True)

# Pastikan semua file memiliki kolom 'title' dan 'genre' untuk digabungkan
if 'genre' not in netflix_data.columns or 'genre' not in demographic_data.columns or 'genre' not in filmtv_data.columns:
    raise ValueError("Kolom 'genre' tidak ditemukan dalam salah satu dataset.")
if 'title' not in netflix_data.columns or 'title' not in demographic_data.columns or 'title' not in filmtv_data.columns:
    raise ValueError("Kolom 'title' tidak ditemukan dalam salah satu dataset.")

# Menggabungkan data
combined_data = pd.concat([netflix_data[['title', 'genre']],
                           demographic_data[['title', 'genre']],
                           filmtv_data[['title', 'genre']]], ignore_index=True)
combined_data['genre'] = combined_data['genre'].str.lower()

# Fungsi untuk mencari film berdasarkan genre
def search_movies_by_genres(genres):
    # Membuat list genre yang diinputkan pengguna menjadi huruf kecil
    genres = [genre.lower().strip() for genre in genres]

    # Filter data berdasarkan genre yang cocok
    matches = pd.DataFrame()
    for genre in genres:
        matches = pd.concat([
            matches,
            combined_data[combined_data['genre'].str.contains(genre, case=False, na=False)]
        ]).drop_duplicates()

    # Mengurutkan hasil berdasarkan judul
    return matches.sort_values('title')

# Input genre dari pengguna
user_input = input("Masukkan genre yang ingin dicari (pisahkan dengan koma, contoh: action, comedy): ")
genre_list = [genre.strip() for genre in user_input.split(',')]

# Mencari dan menampilkan hasil
search_result = search_movies_by_genres(genre_list)

# Menampilkan hasil
if not search_result.empty:
    print(f"\nDitemukan {len(search_result)} film dengan genre yang dicari:")
    for index, row in search_result.iterrows():
        print(f"\nJudul: {row['title']}")
        print(f"Genre: {row['genre']}")
    print(f"\nTotal film ditemukan: {len(search_result)}")
else:
    print("\nTidak ada film yang ditemukan dengan genre tersebut.")
```

```python
# Fungsi rekursif untuk mencari film berdasarkan genre
def search_movies_by_genres_recursively(genres, matches=None, combined_data=None):
    # Inisialisasi parameter untuk iterasi pertama
    if matches is None:
        matches = pd.DataFrame()
    if not genres:  # Basis rekursi: Jika tidak ada genre tersisa, kembalikan hasil
        return matches.drop_duplicates().sort_values('title')

    # Ambil genre pertama dari daftar
    current_genre = genres[0]

    # Cari film yang sesuai dengan genre
    genre_matches = combined_data[combined_data['genre'].str.contains(current_genre, case=False, na=False)]
    matches = pd.concat([matches, genre_matches])  # Gabungkan hasil pencarian saat ini ke `matches`

    # Panggil fungsi secara rekursif dengan sisa genre
    return search_movies_by_genres_recursively(genres[1:], matches, combined_data)


# Input genre dari pengguna
user_input = input("Masukkan genre yang ingin dicari (pisahkan dengan koma, contoh: action, comedy): ")
genre_list = [genre.lower().strip() for genre in user_input.split(',')]  # Format input menjadi huruf kecil

# Mencari film berdasarkan daftar genre yang diberikan menggunakan fungsi rekursif
search_result = search_movies_by_genres_recursively(genre_list, combined_data=combined_data)

# Menampilkan hasil
if not search_result.empty:
    print(f"\nDitemukan {len(search_result)} film dengan genre yang dicari:")
    for index, row in search_result.iterrows():
        print(f"\nJudul: {row['title']}")
        print(f"Genre: {row['genre']}")
    print(f"\nTotal film ditemukan: {len(search_result)}")
else:
    print("\nTidak ada film yang ditemukan dengan genre tersebut.")
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset dengan penanganan error
try:
    netflix_data = pd.read_csv('/content/NetflixOriginals.csv', encoding='latin1', on_bad_lines='skip')
    demographic_data = pd.read_csv('/content/demographic.csv', encoding='latin1', on_bad_lines='skip')
    filmtv_data = pd.read_csv('/content/filmtv_movies.csv', encoding='latin1', on_bad_lines='skip')
except pd.errors.ParserError as e:
    print("Error saat membaca file CSV:", e)

# Penyesuaian kolom
if 'genre' not in filmtv_data.columns:
    filmtv_data['genre'] = 'Unknown'
elif 'genres' in filmtv_data.columns:
    filmtv_data.rename(columns={'genres': 'genre'}, inplace=True)

if 'Genre' in netflix_data.columns:
    netflix_data.rename(columns={'Genre': 'genre'}, inplace=True)
if 'Title' in netflix_data.columns:
    netflix_data.rename(columns={'Title': 'title'}, inplace=True)

if 'genres' in demographic_data.columns:
    demographic_data.rename(columns={'genres': 'genre'}, inplace=True)
if 'Title' in demographic_data.columns:
    demographic_data.rename(columns={'Title': 'title'}, inplace=True)

# Gabungkan data
combined_data = pd.concat([
    netflix_data[['title', 'genre']],
    demographic_data[['title', 'genre']],
    filmtv_data[['title', 'genre']]
], ignore_index=True)
combined_data['genre'] = combined_data['genre'].str.lower()

# Fungsi iteratif untuk mencari film berdasarkan genre
def search_movies_by_genres(genres, data):
    genres = [genre.lower().strip() for genre in genres]
    matches = pd.DataFrame()
    for genre in genres:
        matches = pd.concat([
            matches,
            data[data['genre'].str.contains(genre, case=False, na=False)]
        ]).drop_duplicates()
    return matches.sort_values('title')

# Fungsi rekursif untuk mencari film berdasarkan genre
def search_movies_rekursif(genres, combined_data, index=0, matches=pd.DataFrame()):
    if index >= len(genres):
        return matches.sort_values('title').drop_duplicates()
    current_genre = genres[index].lower().strip()
    filtered = combined_data[combined_data['genre'].str.contains(current_genre, case=False, na=False)]
    matches = pd.concat([matches, filtered])
    return search_movies_rekursif(genres, combined_data, index+1, matches)

# Fungsi untuk mengukur performa iteratif dan rekursif
def measure_performance(method, input_sizes, data):
    times = []  # Untuk menyimpan waktu eksekusi
    for input_size in input_sizes:
        # Buat dummy genres untuk simulasi
        dummy_genres = ['action', 'comedy', 'drama', 'horror', 'thriller'][:input_size]

        # Mulai waktu
        start_time = time.time()

        # Jalankan metode pencarian
        if method == "iterative":
            search_movies_by_genres(dummy_genres, data)
        elif method == "recursive":
            search_movies_rekursif(dummy_genres, data)

        # Akhiri waktu
        end_time = time.time()

        execution_time = end_time - start_time
        times.append(execution_time)

    return times

# Ukuran input genre yang akan diuji
input_sizes = [1, 2, 3, 4, 5]  # Jumlah genre yang akan dicari

# Hitung waktu eksekusi untuk iteratif dan rekursif
iterative_times = measure_performance("iterative", input_sizes, combined_data)
recursive_times = measure_performance("recursive", input_sizes, combined_data)

# Membuat tabel perbedaan performa
performance_table = pd.DataFrame({
    "Jumlah Genre": input_sizes,
    "Waktu Iteratif (detik)": iterative_times,
    "Waktu Rekursif (detik)": recursive_times,
    "Perbedaan (Iter - Rek) (detik)": [i - r for i, r in zip(iterative_times, recursive_times)]
})

# Menampilkan tabel
print("\nTabel Perbandingan Performa:\n")
print(performance_table)

# Plot hasil perbandingan dalam satu grafik
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, iterative_times, label="Iteratif", marker='o')
plt.plot(input_sizes, recursive_times, label="Rekursif", marker='s')
plt.title("Perbandingan Performa Iteratif vs Rekursif", fontsize=14)
plt.xlabel("Jumlah Genre yang Dicari", fontsize=12)
plt.ylabel("Waktu Eksekusi (detik)", fontsize=12)
plt.legend()
plt.grid()

# Tampilkan grafik
plt.show()
```
