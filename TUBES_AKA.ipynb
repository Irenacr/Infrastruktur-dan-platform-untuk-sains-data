```python
import pandas as pd

# Membaca data dari file CSV
netflix_data = pd.read_csv('/content/NetflixOriginals.csv', encoding='latin1')
demographic_data = pd.read_csv('/content/demographic.csv', encoding='latin1')
filmtv_data = pd.read_csv('/content/filmtv_movies.csv', encoding='latin1')

# Check 'genre' or 'genres' column in filmtv_data
if 'genre' not in filmtv_data.columns and 'genres' not in filmtv_data.columns:
    filmtv_data['genre'] = 'Unknown'
elif 'genres' in filmtv_data.columns:
    filmtv_data.rename(columns={'genres': 'genre'}, inplace=True)

# Rename columns
if 'Genre' in netflix_data.columns:
    netflix_data.rename(columns={'Genre': 'genre'}, inplace=True)
if 'Title' in netflix_data.columns:
    netflix_data.rename(columns={'Title': 'title'}, inplace=True)
if 'genres' in demographic_data.columns:
    demographic_data.rename(columns={'genres': 'genre'}, inplace=True)
if 'Title' in demographic_data.columns:
    demographic_data.rename(columns={'Title': 'title'}, inplace=True)

# Check required columns
if 'genre' not in netflix_data.columns or 'genre' not in demographic_data.columns or 'genre' not in filmtv_data.columns:
    raise ValueError("Kolom 'genre' tidak ditemukan dalam salah satu dataset.")
if 'title' not in netflix_data.columns or 'title' not in demographic_data.columns or 'title' not in filmtv_data.columns:
    raise ValueError("Kolom 'title' tidak ditemukan dalam salah satu dataset.")

# Combine datasets
combined_data = pd.concat([
    netflix_data[['title', 'genre']],
    demographic_data[['title', 'genre']],
    filmtv_data[['title', 'genre']]
], ignore_index=True)

# Normalize genre column
combined_data['genre'] = combined_data['genre'].str.lower()

# Fungsi rekursif untuk pencarian film berdasarkan genre
def search_movies_rekursif(genres, combined_data, index=0, matches=pd.DataFrame()):
    # Base case: Jika sudah memproses seluruh genre dalam input
    if index >= len(genres):
        return matches.sort_values('title').drop_duplicates()

    # Memproses genre saat ini
    current_genre = genres[index].lower().strip()

    # Filter baris-baris yang mengandung genre terkait
    filtered = combined_data[combined_data['genre'].str.contains(current_genre, case=False, na=False)]

    # Tambahkan hasil ke dalam matches
    matches = pd.concat([matches, filtered])

    # Rekursi untuk genre berikutnya
    return search_movies_rekursif(genres, combined_data, index + 1, matches)

# Input genre dari pengguna
user_input = input("Masukkan genre yang ingin dicari (pisahkan dengan koma, contoh: action, comedy): ")
genre_list = [genre.strip() for genre in user_input.split(',')]

# Mencari dan menampilkan hasil
search_result = search_movies_rekursif(genre_list, combined_data)

if not search_result.empty:
    print(f"\nDitemukan {len(search_result)} film dengan genre yang dicari:")
    # Menampilkan hasil dengan format rapi
    for index, row in search_result.iterrows():
        print(f"\nJudul: {row['title']}")
        print(f"Genre: {row['genre']}")
    print(f"\nTotal film ditemukan: {len(search_result)}")
else:
    print("\nTidak ada film yang ditemukan dengan genre tersebut.")
```
